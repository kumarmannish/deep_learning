# Transpose

```python
temp = np.array([[1, 2, 3, 4]])
print(temp)
print(type(temp))
print(temp.transpose())
print(temp.T)
temp_t = temp.T
print(temp_t.T)
```

```python
temp = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
print(temp)
print(type(temp))
print(temp.transpose())
print(temp.T)
temp_t = temp.T
print(temp_t.T)
```

```python
temp = torch.tensor([[1, 2, 3, 4]])
print(temp)
print(type(temp))
print(temp.T)
temp_t = temp.T
print(temp_t)
print(temp_t.T)
```

```python
temp = torch.tensor([[1, 2, 3, 4], [5, 6, 7, 8]])
print(temp)
print(type(temp))
print(temp.T)
temp_t = temp.T
print(temp_t)
print(temp_t.T)
```

# Dot Product

```python
temp_1 = np.array([1, 2, 3, 4])
temp_2 = np.array([5, 6, 7, 8])
print(np.dot(temp_1, temp_2))
print(np.sum(temp_1*temp_2))
```

```python
temp_1 = torch.tensor([1, 2, 3, 4])
temp_2 = torch.tensor([5, 6, 7, 8])
print(torch.dot(temp_1, temp_2))
print(torch.sum(temp_1*temp_2))
```

# Softmax Function

```python
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
temp = [1, 2, 3]
temp_exp = np.exp(temp)
sum_exp = np.sum(temp_exp)

sigma = temp_exp/sum_exp
print(sigma)

plt.plot(temp, sigma, "ko")
plt.xlabel("Original Number (temp)")
plt.ylabel("Softmax Number (sigma)")
```

```python
temp = np.arange(-10, 30, 4)
exp_temp = np.exp(temp)
sum_exp = np.sum(exp_temp)

sigma = exp_temp/sum_exp
print(sigma)
print(np.sum(sigma))

plt.plot(temp, sigma, "ko")
plt.xlabel("Original Number (temp)")
plt.ylabel("Softmax Number (sigma)")
```

```python
softmax = nn.Softmax(dim=0)

sigma = softmax(torch.Tensor([1, 2, 3]))

print(sigma)
```

# Logarithms Function

```python
x = np.linspace(0.0002, 1, 200)
logx = np.log(x)

plt.figure(figsize=(5, 3), dpi=200)
plt.plot(x, logx, 'ks-', markerfacecolor='w')
plt.xlabel("x")
plt.ylabel("log(x)")
```

```python
x = np.linspace(0.0002, 1, 200)

logx = np.log(x)
expx = np.exp(x)

plt.figure(figsize=(7, 3.5), dpi=300)
plt.plot(x, logx, 'ks-', markerfacecolor='w')
plt.plot(x, expx, 'ks-', markerfacecolor='w')
plt.xlabel("x")
plt.ylabel("f(x)")
```

```python
x = np.linspace(0.0002, 1, 200)

logx = np.log(x)
expx = np.exp(x)

plt.figure(figsize=(7, 3.5), dpi=300)
plt.plot(x, np.exp(logx), 'o', markerfacecolor='w')
plt.plot(x, np.log(expx), 'k', markerfacecolor='w')
plt.xlabel("x")
plt.ylabel("g(f(x))")
```

# Entropy or Binary Entropy

```python
X = [0.25, 0.75]
H = 0
for p in X:
    H -= p*np.log(p)

print(H)
```

# Cross-Entropy or Binary Cross-Entropy

```python
p = [1, 0]
q = [0.25, 0.75]

H = 0
for i in range(len(p)):
    H -= p[i]*np.log(q[i])

print(H)
```

```python
import torch.nn.functional as F
```

```python
p_t = torch.Tensor(p)
q_t = torch.Tensor(q)

F.binary_cross_entropy(q_t, p_t)
```

# Min/Max and ArgMin/ArgMax

```python
x = np.array([1, 2, 3, -1, 0])

print(np.min(x))
print(np.max(x))
print(np.argmin(x))
print(np.argmax(x))
```

```python
x = np.array([[1, 2, 3, 4, -1], 
              [5, 6, 10, -2, 0]])

print(np.min(x, axis=0))
print(np.min(x, axis=1))
print(np.max(x, axis=0))
print(np.max(x, axis=1))

print(np.argmin(x, axis=0))
print(np.argmin(x, axis=1))
print(np.argmax(x, axis=0))
print(np.argmax(x, axis=1))
```

```python
x = torch.Tensor([1, 2, 3, -1, 0])

print(torch.argmin(x))
print(torch.argmax(x))
print(torch.min(x))
print(torch.max(x))
```

```python
x = torch.Tensor([[1, 2, 3, 4, 5], [-1, -2, 10, -4, 2]])

print(torch.min(x, axis=0).values)
print(torch.min(x, axis=0).indices)
print(torch.min(x, axis=1).values)
print(torch.min(x, axis=1).indices)
print(torch.max(x, axis=0).values)
print(torch.max(x, axis=0).indices)
print(torch.max(x, axis=1).values)
print(torch.max(x, axis=1).indices)
```

# Mean and Variance

```python
x = [1, 2, 4, 6, 5, 4, 0]
n = len(x)

mean_1 = np.mean(x)
mean_2 = np.sum(x)/n

print(mean_1, mean_2)
```

```python
var_1 = np.var(x) # In this we're not using the biased variance
var_2 = np.sum((x-mean_1)**2)/(n-1) # In this we're using the unbiased variance so this will remove the bias, it is also related to the degrees of freedom
# We are removing the one degree of freedom

print(var_1, var_2)
```

```python
var_1 = np.var(x, ddof=1)
var_2 = np.sum((x-mean_1)**2)/(n-1)

print(var_1, var_2)
```

# t-test

```python
import scipy.stats as stats
```

```python
num_1 = 30
num_2 = 40
mean_1 = 1
mean_2 = 2

data_1 = np.random.randn(num_1)
data_2 = np.random.randn(num_2)

plt.plot(np.zeros(num_1), data_1, 'ro', markerfacecolor='w', markersize=14)
plt.plot(np.ones(num_2), data_2, 'bs', markerfacecolor='w', markersize=14)
plt.xlim([-1, 2])
plt.xticks([0, 1], labels=["Group 1", "Group 2"])
```

```python
t, p = stats.ttest_ind(data_1, data_2)
print(t, p)
```
